```python

years = [1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]

def get_FA_books(year=2020):
  names_links = []
  url = 'https://www.foreignaffairs.com/issues/archive/' + str(year)
  headers = {'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:80.0) Gecko/20100101 Firefox/80.0'}
  soup = BeautifulSoup(requests.get(url, headers=headers).content, 'html.parser')
  class_name = 'views-field views-field-title'
  items = soup.find_all(class_=class_name)
  prefix_url = 'https://www.foreignaffairs.com'
  for it in items:
    name = it.get_text().strip()
    a_href = it.find('a')
    href = a_href.get('href')
    names_links.append( (name, prefix_url+href) )
    #print(name, names_links[name]) 
  return names_links

def get_FA_one_book(book_url):
  url = book_url
  titles_links = []
  prefix_url = 'https://www.foreignaffairs.com'
  headers = {'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:80.0) Gecko/20100101 Firefox/80.0'}
  soup = BeautifulSoup(requests.get(url, headers=headers).content, 'html.parser')
  class_name = 'views-field views-field-title'
  items = soup.find_all(class_=class_name)
  for it in items:
    title = it.get_text().strip()
    a_href = it.find('a')
    href = a_href.get('href')
    #print(name, '\t' ,prefix_url+href)
    titles_links.append( (title, prefix_url+href) )
  return titles_links

def get_FA_one_page(page_url):
  url = page_url
  headers = {'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:80.0) Gecko/20100101 Firefox/80.0'}
  soup = BeautifulSoup(requests.get(url, headers=headers).content, 'html.parser')
  #print(soup)
  # title
  try:
    title = soup.find(class_='f-serif ls-0 article-title pt-2').get_text().strip()
  except:
    title = soup.find(class_='f-serif ls-0 article-title pt-2 mb-4').get_text().strip()
  # content
  try:
    content = soup.find(class_='article-dropcap--inner paywall-content').get_text().strip() # 
  except:
    content = soup.find(class_='article-dropcap ls-0 f-serif').get_text().strip()
  #items = soup.find_all(class_=class_name)
  #for it in items:
  #  print(it.get_text())
  lns = []
  for ln in content.split('\n'):
    if len(ln.strip())>0:
      lns.append(ln.strip())
  first_paragraph = lns[0].strip()
  return title, first_paragraph


###
# book_url = 'https://www.foreignaffairs.com/issues/2019/98/6'
def get_one_journal_data(book_url):
  journal = []
  titles_links = get_FA_one_book(book_url)
    # There Will Be a One-State Solution 	 https://www.foreignaffairs.com/articles/israel/2019-10-15/there-will-be-one-state-solution
    # 'The Virtue of Monopoly'  'https://www.foreignaffairs.com/reviews/virtue-monopoly'
    # 'Two Books on Latin American History' 'https://www.foreignaffairs.com/reviews/capsule-review/2019-10-15/two-books-latin-american-history'
    # 'Trumpâ€™s Middle East' 'https://www.foreignaffairs.com/issue-packages/2019-10-15/trumps-middle-east'
  for it2 in titles_links:
    title1, page_url = it2
    #print( title1, page_url )
    #page_url = 'https://www.foreignaffairs.com/articles/israel/2019-10-15/there-will-be-one-state-solution'
    title2, first_paragraph = get_FA_one_page(page_url)  
    #print( title2, first_paragraph )
    journal.append( (title1, page_url, title2, first_paragraph) )
  return journal

year = 2010
names_links = get_FA_books(year)
  # 'November/December 2019': 'https://www.foreignaffairs.com/issues/2019/98/6'
for it1 in names_links[:]: # every 2 month, 
  time_publish, book_url = it1
  file_name = time_publish.replace('/', ' ') 
  file_name = file_name.replace(' ', '_') + ".csv"
  journal = get_one_journal_data(book_url)
  print( file_name, time_publish, book_url, len(journal) )
  col_names = ['title1', 'page_url', 'title2', 'first_paragraph']
  df_data = pd.DataFrame(journal, columns=col_names)
  df_data.to_csv(file_name)
```
